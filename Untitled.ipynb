```python
def show_sensitivity(sbj_id, cov=True, sensitivity=True):
    sbj_path = path.join(config.data_path, config.map_subjects[sbj_id][0])
    
    if cov==True:
        dirname=path.join(sbj_path, 'Figures')
        for filename in os.listdir(dirname):
            root, ext = os.path.splitext(filename)
            if root.startswith("covariance") and root.endswith("_matrix"):
                display(Image.open(path.join(sbj_path, 'Figures', filename)))
                strings = root.split("_")[:2]
                eig = '_'.join(strings) + "_eigenvalue_index.png"
                display(Image.open(path.join(sbj_path, 'Figures', eig)))
    if sensitivity==True:
        display(Image.open(path.join(sbj_path, 'Figures', 'leadfield_solved.png')))
        display(Image.open(path.join(sbj_path, 'Figures', 'Normal Orientation Sensitivity_solved.png')))
        display(Image.open(path.join(sbj_path, 'Figures', 'Gradiometers Sensitivity_solved.png')))
        display(Image.open(path.join(sbj_path, 'Figures', 'Source depth_solved.png')))
```
