```python
leadfield = fwd_eegmeg['sol']['data']

picks_eeg = mne.pick_types(fwd_eegmeg['info'], meg=False, eeg=True)
picks_grad = mne.pick_types(fwd_eegmeg['info'], meg='grad', eeg=False)
picks_meg = mne.pick_types(fwd_eegmeg['info'], meg='mag', eeg=False)
```

```python
fig, axx = plt.subplots(3,1, figsize=(20, 8))
[sns.heatmap(leadfield[picks, :], robust=True, ax=ax, cbar=False) for picks, ax in zip([picks_eeg, picks_grad, picks_meg], axx.ravel())];
```

```python
fig, ax = plt.subplots(3,1, tight_layout=True, figsize=(8, 16))
for k, p in enumerate([picks_eeg, picks_grad, picks_meg]):
    sns.heatmap(np.cov(leadfield[p, :]), ax=ax[ k], square=True, cmap='viridis')
    #sns.heatmap(np.cov(leadfield[p, :].T), ax=ax[1, k])
```
